### 1.1. Data Numerik
Data yang bersifat numerik dapat diberikan nilai mean, median, modus, standar deviasi, variansi, range, nilai minimum, maksimum, kuartil, IQR, skewness dan kurtosis.

````python
#sorting kolom numerik
datanum = data.iloc[:, 1:].select_dtypes(include=[np.number]) 

# count, mean, std, min, quartile 1(25%), quartile 2(50%), quartile 3(75%), max 
descnum_statistic = datanum.describe()

# median
descnum_statistic.loc['median'] = datanum.median()

# modus
descnum_statistic.loc['modus'] = datanum.mode().iloc[0] 

# variansi
descnum_statistic.loc['variansi'] = datanum.var()

# range
descnum_statistic.loc['range'] = datanum.max() - datanum.min()

# skewness
descnum_statistic.loc['skewness'] = datanum.skew()

# kurtosis
descnum_statistic.loc['kurtosis'] = datanum.kurt()

# Display
descnum_statistic
````

## 2.2 Income Column

````python
fig = plt.figure(figsize=(10,5))
sns.boxplot(data.Income)
plt.title('Box Plot: Income', fontsize=15)
plt.show()

IQR = datanum.Income.quantile(0.75) - datanum.Income.quantile(0.25)
upper_limit = datanum.Income.quantile(0.75) + 1.5 * IQR
lower_limit = datanum.Income.quantile(0.25) - 1.5 * IQR

# outlier
display(data[~((data.Income < upper_limit) & (data.Income > lower_limit))])

# penanganan
percentile_5 = data.Income.quantile(0.05)
percentile_95 = data.Income.quantile(0.95)

# Winsorization 90%
data['Income_winsorized'] = data['Income'].clip(percentile_5, percentile_95)

fig = plt.figure(figsize=(10,5))
sns.boxplot(data.Income_winsorized)
plt.title('Box Plot: Income (Winsorized)', fontsize=15)
plt.show()
````
